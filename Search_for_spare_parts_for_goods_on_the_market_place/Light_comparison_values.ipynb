# Comparison values. Поиск одинаковых товаров на маркет плейсе

```python
import pandas as pd
df = pd.read_json('2108.json')
```

```python
search = input('Поиск:')
indxs = (df[df["product_name"].str.lower().str.contains(search)]
         .groupby(['product_name', 'name_store']).first().reset_index()
         ['_id'].tolist())
df.query('_id in @indxs')

# например: батончик злаковый fitness
```
