```python
import pandas as pd

labels = ['sex','height','weight','bmi',
          'bsa','age','diagnosis',
          'medication_main','medications_taken',
          'days_in_hospital','fv','lg_kdr_size',
          'lg_hipoplaz','pg_big','pg_hypertrop',
          'dmgp','ekg_sinus','ekg_eos','ekg_qrs',
          'ekg_qrst','ekg_pq','total']
types = ['int64','float64','float64','float64','float64',
         'float64','object','object','object','int64',
         'float64','float64','int64','int64','int64',
         'int64','int64','int64','float64','float64',
         'float64','int64']

data = pd.DataFrame()
i = 0
while i<len(labels):
    value = input(f'{labels[i]}: ')
    try:
        data[labels[i]] = pd.Series([value]).astype(types[i])
        i += 1
    except:
        print(f'Неверный формат для {labels[i]}! Важны детали. Введите в стиле float, int или str. Пример в ячейке выше!')
        
display(data)
```

## Модель

```python
import pandas as pd
from joblib import dump, load
model = load('model.joblib')

answer = round(model.predict_proba(data)[0][1]*100)
print('Вероятность врожденного порога сердца у ребенка: {}%'.format(answer))
```
